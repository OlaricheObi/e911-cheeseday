# Import Packages

```python
from shapely.geometry import Point
import pandas as pd
import datetime
import time
import requests
import geopandas as gpd
import json
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
import folium.plugins as plugins
from folium import plugins
from folium.plugins import HeatMap
from folium.plugins import HeatMapWithTime
```

```python
#Import CSV
derecho_calls = pd.read_csv('../data/911_Phone Calls_Derecho_050320_050420.csv') 

#Rename Columns
derecho_calls = derecho_calls.rename(columns = {'Seizure DateTime': 'time', 'CallTypeId': 'call', 'ALI Latitude': 'lat', 'ALI Longitude': 'lng', 'Cell Tower Address': 'tower_addy' })

#Convert time column to datetime
derecho_calls.time = pd.to_datetime(derecho_calls.time)
```

```python
derecho_calls.info()
```

```python
#Exclude NaN values
derecho_calls.dropna(inplace= True)
```

```python
derecho_calls.head()
```

```python
#create the basemap
heatmap = folium.Map(location = [36.1627, -86.7816], 
                    zoom_start = 10)

derecho_calls['count'] = 1 

derecho_calls['minute'] = pd.DatetimeIndex(derecho_calls['time']).minute
derecho_calls['second'] = pd.DatetimeIndex(derecho_calls['time']).second
derecho_calls['hour'] = pd.DatetimeIndex(derecho_calls['time']).hour

df = pd.DataFrame(
    derecho_calls.groupby(['lat', 'lng'])\
      ['count'].sum().sort_values(ascending=False))

lst = df.groupby(
    ['lat', 'lng'])\
    .sum().reset_index().values.tolist()
```

```python
df
```

```python
HeatMap(data = lst, radius = 20).add_to(heatmap);
```

```python
derecho_list = []
for i in derecho2['time2'].unique():
    weight=[]
    for index, instance in derecho2[derecho2['time2'] == i].iterrows():
        weight.append([instance['lat'],instance['lng']])
    derecho_list.append(weight)
```

```python
#creating a time index
timestamp = []
for i in derecho2['time2'].unique():
    timestamp.append(i)

#formatting the index
dates = [pd.to_datetime(d).strftime('%d/%m/%Y, %H:%M:%S') for d in timestamp]
```

```python
anothermap = folium.Map(location = [36.1627, -86.7816], 
                    zoom_start = 10, tiles = 'cartodbdark_matter')
```

```python
HeatMapWithTime(derecho_list ,radius=30, auto_play=True, position='bottomright', name="cluster", index=dates, max_opacity=0.9).add_to(anothermap)
anothermap
```
